**Describe your state and actions:**
  <br/>  I set the state to a list of the path the algorithm has currently chosen to be evaluated. For actions, I chose the set of all possible swaps between any two cities in the path. For example, if the algorithm was currently evaluating the list ["city1", "city2", "city3"], a possible swap would be "city1" and "city3". If the algorithm chose that, the resulting path would be ["city3", "city2", "city1"].  

```python



from search import Problem, hill_climbing, simulated_annealing, \
    exp_schedule, genetic_search
from random import randrange
import math
from statistics import mean 
from itertools import combinations 
import copy
import time

class TSP(Problem):
    '''
        state: currently selected path
        actions: Any swap that cn be done between any two cities 
    '''

    def __init__(self, adj_matrix, initial):
        self.adj_matrix = adj_matrix
        self.initial = initial
        
    def actions(self, state):
        #returns a list of every possible swap of the two cities
        return list([city for city in combinations(state, 2)])
    
    '''returns the result of the selected move
    (swapping two selected cities in that 
     selected solution path order) '''
    def result(self, state, move):
        #deep copy as to not directy modify the current state
        next_state = copy.deepcopy(state) 
        city1_index = state.index(move[0])
        city2_index = state.index(move[1])
        next_state[city1_index], next_state[city2_index]= next_state[city2_index], next_state[city1_index]
        return next_state
    
    # computes the distance of the current
    # selected path
    def value(self, state):
        #calculates the last 
        score = self.adj_matrix[state[len(state)-1]] [state[0]]
        for i in range(len(state)-2):
            score += self.adj_matrix[state[i]] [state[i+1]] 
            
        return (-1 * score) # algorithms want to maximize score, so return negatve
    
```
